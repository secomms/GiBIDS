## All

```python
import matplotlib.pyplot as plt
n = 16

col_weights = 'black'
col_unweights = 'black'
col_mix = 'black'
line_style_weights = 'dashed'
line_style_unweights = 'dotted'
line_style_mix = 'dashdot'


fig, axs = plt.subplots(1, 3, figsize=(12, 5))

# first image
axs[0].plot([1, 2, 3, 4, 5, 6, 7, 8], sigma_1000_unweighted, linewidth=line_width,
            linestyle=line_style_unweights, c=col_unweights, label='Unweighted case')
axs[0].plot([1, 2, 3, 4, 5, 6, 7, 8], sigma_1000_weighted, linewidth=line_width,
            linestyle=line_style_weights, c=col_weights, label='Weighted case')
axs[0].plot([1, 2, 3, 4, 5, 6, 7, 8], sigma_1000_mix_weighted, linewidth=line_width,
            linestyle=line_style_mix, c=col_mix, label='Mixed case')
axs[0].set_ylabel(yLabel, fontsize=n)
axs[0].set_title('\u03C3$^1$', fontsize=n)
axs[0].legend(fontsize=n)
axs[0].grid()
axs[0].tick_params(axis='both', labelsize=n)

# second image
axs[1].plot([1, 2, 3, 4, 5, 6, 7, 8], sigma_5000_unweighted, linewidth=line_width,
            linestyle=line_style_unweights, c=col_unweights, label='Unweighted case')
axs[1].plot([1, 2, 3, 4, 5, 6, 7, 8], sigma_5000_weighted, linewidth=line_width,
            linestyle=line_style_weights, c=col_weights, label='Weighted case')
axs[1].plot([1, 2, 3, 4, 5, 6, 7, 8], sigma_5000_mix_weighted, linewidth=line_width,
            linestyle=line_style_mix, c=col_mix, label='Mixed case')
axs[1].set_title('\u03C3$^5$', fontsize=n)
axs[1].legend(fontsize=n)
axs[1].grid(which='both', axis='both') 
axs[1].tick_params(axis='both', labelsize=n)
axs[1].set_xlabel(xLabel, fontsize=n)

# third image
axs[2].plot([1, 2, 3, 4, 5, 6, 7, 8], sigma_Max_unweighted, linewidth=line_width,
            linestyle=line_style_unweights, c=col_unweights, label='Unweighted case')
axs[2].plot([1, 2, 3, 4, 5, 6, 7, 8], sigma_Max_weighted, linewidth=line_width,
            linestyle=line_style_weights, c=col_weights, label='Weighted case')
axs[2].plot([1, 2, 3, 4, 5, 6, 7, 8], sigma_Max_mix_weighted, linewidth=line_width,
            linestyle=line_style_mix, c=col_mix, label='Mixed case')
axs[2].set_title('\u03C3$^N$', fontsize=n)
axs[2].legend(fontsize=n)
axs[2].grid(which='both', axis='both') 
axs[2].tick_params(axis='both', labelsize=n)

# Remove y-axis tick labels for the second and third subplots
axs[1].tick_params(axis='y', labelleft=False)
axs[2].tick_params(axis='y', labelleft=False)

# Adjust y-axis limits for uniform scaling
y_min = min(
    min(sigma_1000_unweighted), min(sigma_1000_weighted), min(sigma_1000_mix_weighted),
    min(sigma_5000_unweighted), min(sigma_5000_weighted), min(sigma_5000_mix_weighted),
    min(sigma_Max_unweighted), min(sigma_Max_weighted), min(sigma_Max_mix_weighted)
)
y_max = max(
    max(sigma_1000_unweighted), max(sigma_1000_weighted), max(sigma_1000_mix_weighted),
    max(sigma_5000_unweighted), max(sigma_5000_weighted), max(sigma_5000_mix_weighted),
    max(sigma_Max_unweighted), max(sigma_Max_weighted), max(sigma_Max_mix_weighted)
)

for ax in axs:
    ax.set_ylim(y_min - 0.0005, y_max + 0.0005)


plt.tight_layout()
plt.show()
fig.savefig('fs.pdf', format='pdf', dpi=300)
```
